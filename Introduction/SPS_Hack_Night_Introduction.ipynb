## If/Else Statements

Adding simple control structures

```
A = 10
B = 5
if A >= 5:
   C = A - B
else:
   C = A + B
if A == 10:
   A = A + 1
print(C)
print(A)
```

## Loops

```
for i in range(5):
   print(i)

x = 3
A = 10
while x < A:
   x = 2 * x

print(x)
```

## Functions
A cleaner way to repeat a task.

```
def factorial(x):
   if x == 1:
       return 1
   else:
       return x*factorial(x-1)

print(factorial(3))
print(factorial(10))
```

## Arrays and Dictionaries
Storing goups of variables in an organized fashion.

```
A = [0, 1, 2]
B = {
   "name": "Student",
   "class_year": 2021
}

for i in range(len(A)):
   print(A[i])
for i in B:
   print(B[i])

for i in A:
   print(i)
for i in B:
   print(i)
```
