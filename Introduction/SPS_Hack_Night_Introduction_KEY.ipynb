## Functions
A cleaner way to repeat a task.

```
def factorial(x):
   if x == 1:
       return 1
   else:
       return x*factorial(x-1)

print(factorial(3))
print(factorial(10))
```

## Newtonian Gravity

Let's analytically solve for an object in orbit around the Earth at (0,0) as a 2D system.

```
# Gravity Solver
# Noah Paladino
# Sept 2020

# Import Libraries
import math
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Initial Conditions
r = [6756e3, 0]    # Position (x,y); [m]
v = [0,7.66e3]     # Velocity (x,y); [m/s]
m = 10000.       # Mass [kg]

# Constants
G = 6.7e-11   # Newton's Gravitational Constant [N*km^2/kg^2]
M = 6.0e24    # Mass of Earth [kg]
delta_t = 0.1  # Seconds per increment

def F_g(position, mass):
    distance = math.sqrt(position[0]**2 + position[1]**2)
    return [-(G*M*mass/(distance**3))*position[0], -(G*M*mass/(distance**3))*position[1]]

# Store Position History
x_list = [r[0]] 
y_list = [r[1]] 
print(r)
# Update State
for i in range(10000000):
    f = F_g(r,m)            # Get Force
    a = [f[0]/m, f[1]/m]    # Get Acceleration from Force

    # print("R="+str(r)+"  F="+str(f)+"  D="+str(math.sqrt(r[0]**2 + r[1]**2)))

    # Update Position
    r[0] = r[0] + v[0]*delta_t + 0.5*a[0]*(delta_t**2.)
    r[1] = r[1] + v[1]*delta_t + 0.5*a[1]*(delta_t**2.)

    # Update Velocity
    v[0] = v[0] + a[0]*delta_t
    v[1] = v[1] + a[1]*delta_t

    # Update Position History
    x_list.append(r[0])
    y_list.append(r[1])
    
    

# Plot
plt.plot(x_list,y_list)
plt.show()

```
